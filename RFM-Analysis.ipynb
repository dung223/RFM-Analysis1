```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import squarify
df=pd.read_excel("C:\\Users\\maitr\\Desktop\\data\\Python\\Final projcet\\Final_project_RFM\\ecommerce_retail.xlsx")
df
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df=df.dropna(subset=['CustomerID'],axis=0).drop_duplicates()
df
```

```python
day= datetime.datetime.strptime('2011-12-31', '%Y-%m-%d')
day
```

```python
df1 = df.query('UnitPrice > 0 and Quantity>0')
df1
```

```python
df2=df1[df1['Country']=='United Kingdom']
df2
```

```python
df2['Revenue']=df2['Quantity'] * df2['UnitPrice']
df2
```

```python
df3 = df2.groupby(['InvoiceNo', 'CustomerID']).agg({'InvoiceDate': 'max','Revenue': 'sum'}).reset_index()
df3
```

```python
rfm = df3.groupby('CustomerID').agg({'InvoiceDate': 'max', 'InvoiceNo': 'nunique', 'Revenue': 'sum'}).reset_index()
rfm['Recency']=(day-rfm['InvoiceDate']).dt.days
rfm = rfm.drop('InvoiceDate', axis=1)
rfm
```

```python
rfm.columns=['CustomerID','Frequency','Monetary','Recency']
rfm
```

```python
r_labels, f_labels, m_labels = range(5, 0, -1), range(1, 6), range(1, 6)

rfm['R_score'] = pd.qcut(rfm['Recency'], q=5, labels=r_labels)
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=5, labels=f_labels)
rfm['M_score'] = pd.qcut(rfm['Monetary'], q=5, labels=m_labels)

rfm
```

```python
rfm['RFM Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm['RFM Score'] = rfm['RFM Score'].astype(int)
rfm
```

```python
seg=pd.read_excel("C:\\Users\\maitr\\Desktop\\data\\Python\\Final projcet\\Final_project_RFM\ecommerce_retail.xlsx",sheet_name="Segmentation")
seg
```

```python
seg['RFM Score'] = seg['RFM Score'].str.split(',')
seg= seg.explode('RFM Score').reset_index(drop=True)
seg['RFM Score']=seg['RFM Score'].astype(int)
seg
```

```python
final_df=rfm.merge(seg,on='RFM Score', how='left')
final_df
```

```python
colnames = ['Recency', 'Frequency', 'Monetary']
for col in colnames:
    fig, ax = plt.subplots(figsize=(12,3))
    sns.distplot(final_df[col])
    ax.set_title('Distribution of %s' % col)
    plt.show()
```

```python
ex_df = final_df.groupby('Segment').agg({'CustomerID': 'count',
                                     'Monetary': 'sum',
                                     'Recency': 'mean'}).reset_index()
ex_df.columns = ['Segment', 'Cust_count', 'Total_monetary', 'Avg_recency']
ex_df.sort_values(by='Cust_count', ascending=False, inplace=True)
ex_df
```

```python
ex_df['Cus_share']=ex_df['Cust_count']/ex_df['Cust_count'].sum()
ex_df['Monetary_share'] = ex_df['Total_monetary'] / ex_df['Total_monetary'].sum()
ex_df
```

```python
colors = ['#FF0000','#808080','#800080','#00FFFF','#7FFF00','#D2691E','#006400','#FF00FF','#FFD700','#808000','#FFC0CB']
fig, ax = plt.subplots(1, figsize = (16,8))
squarify.plot(sizes=ex_df['Cust_count'],
              label=ex_df['Segment'],
              value=[f'{x*100:.2f}%' for x in ex_df['Cus_share']],
              alpha=.8,
              color=colors)
plt.title('RFM Segments of Customer Count', fontsize=16)
plt.axis('off')
plt.show()
```

```python
colors = ['#FF0000','#808080','#800080','#00FFFF','#7FFF00','#D2691E','#006400','#FF00FF','#FFD700','#808000','#FFC0CB']
fig, ax = plt.subplots(1, figsize = (16,8))
squarify.plot(sizes=ex_df['Monetary_share'],
              label=ex_df['Segment'],
              value=[f'{x*100:.2f}%' for x in ex_df['Monetary_share']],
              alpha=.8,
              color=colors)
             
plt.title('RFM Segments of Total Sales', fontsize=16)
plt.axis('off')
plt.show()
```
